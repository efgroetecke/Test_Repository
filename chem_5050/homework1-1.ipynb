```python
## Chem 5050 Homework 1
## Problem 1 Part A

## import needed packages
import pandas as pd

## import the volume and pressure data csv file as a pandas dataframe
df = pd.read_csv(r"C:\Users\egroe\Downloads\volume_pressure_data.csv")

## filter the dataframe to print only the first few rows
filtered_df = df[(df['Volume'] > 39)]
print(filtered_df)

## calculate and print the maximum volume (column 1 value) with units
max_vol = df['Volume'].max()
print("The maximum volume is" , max_vol , "cubic inches")

## calculate and print the maximum pressure (column 2 value) with units
max_pres = df['Pressure'].max()
print("The maximum pressure is" , max_pres , "Hg")

## Problem 1 Part B

## import needed packages
import matplotlib
import matplotlib.pyplot as plt

## set x and y values to the volume and pressure data in the dataframe
x = df['Volume']
y = df['Pressure']

## generate a scatter plot visualizing the set x and y values with labels and a grid
plt.scatter(x, y, color = 'blue')
plt.xlabel('Volume (cubic inches)')
plt.ylabel('Pressure (in Hg)')
plt.title('Relationship Between Pressure and Volume')
plt.grid('both')
plt.show()

## Problem 1 Part C

## import needed packages
import numpy as np

## fit a quadratic curve to the scatter plot data
## also fit a higher-order polynomial for a better fit to the scatter plot data (5th degree)
## I fitted two curves here to show the better fit that a higher-order polynomial provides while retaining the data for the quadratic
coef = np.polyfit(x, y, 2)
coef_5 = np.polyfit(x, y, 5)

## generate x values used to plot the quadratic and 5th degree polynomial
## generate 100 points between the minimum and maximum x values
x_new = np.linspace(min(x), max(x), 25)

## generate y values used to plot the quadratic and 5th degree polynomial
y_fitted = np.polyval(coef, x_new)
y_fitted_5 = np.polyval(coef_5, x_new)

## plot the scatter plot data with the new fitted polynomial
plt.scatter(x, y, color = 'blue', label = 'Original Data')
plt.plot(x_new, y_fitted, color = 'red', label = 'Fitted Polynomial (Degree 2)')
plt.plot(x_new, y_fitted_5, color = 'orange', label = 'Fitted Polynomial (Degree 5)')
plt.xlabel('Volume (cubic inches)')
plt.ylabel('Pressure (in Hg)')
plt.title('Relationship Between Pressure and Volume: Fitted')
plt.legend()
plt.grid('both')
plt.show()

## Problem 1 Part D

## print the coefficients for the quadratic
print('The coefficients for the quadratic are', coef)

## print the coefficients for the 5th degree polynomial
print('The coefficients for the 5th degree polynomial are', coef_5)

## The relationship shown in the plot between pressure and volume makes sense, as Boyle's law indicates that the two variables have an inverse (parabolic) relationship

## Problem 1 Part E

## calculate the RMSE between the quadratic fitting and the data
rmse_quad = np.sqrt(np.mean((y - y_fitted) ** 2))
print(rmse_quad)

## calculate the RMSE between the 5th degree fitting at the data
rmse_5 = np.sqrt(np.mean(y - y_fitted_5) ** 2)
print(rmse_5)

## The quadratic fitting is a relatively poor fitting to the data; it has an RMSE of approx 48. This indicates relatively significant deviation between the actual values and the predicted values.
## The 5th degree polynomial is well fit to the data; it has an RMSE of approx 6.5. This indicates that the predicted values are similar to the actual values.
## This makes sense because a polynomial with more terms is more precise, and can better adapt to the data. 
```
